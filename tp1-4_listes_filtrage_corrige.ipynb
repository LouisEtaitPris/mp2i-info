# Filtrage
**<p style = "color:purple">Exercice :</p>** 
1. Ecrire une fonction `voyelle : char -> bool` qui détermine si un caractère est une voyelle ou non.
1. En déduire une fonction `nb_voy : string -> int` qui renvoie le nombre de voyelles présentes dans un mot.
1. Lever une exception si le mot passé en argument comporte un espace.
<hr>

````ocaml
(* 1 *)
let voyelle = function
    | 'a' | 'e' | 'i' | 'o' | 'u' | 'y' -> true
    | _ -> false
;;
````

````ocaml
(* 2 *)
let nb_voy mot = 
    let compte = ref 0 in
    let nb_lettres = String.length mot in
    for lettre = 0 to nb_lettres - 1 do
        if voyelle mot.[lettre] then compte := !compte + 1
    done;
!compte;;

assert (nb_voy "scrabble" = 2) ;
assert (nb_voy "evaluation" = 6);;
````

````ocaml
(* 3 *)
exception Espace;;
let nb_voy mot = 
    let compte = ref 0 in
    let nb_lettres = String.length mot in
    for lettre = 0 to nb_lettres - 1 do
        if mot.[lettre] = ' ' then raise Espace 
        else if voyelle mot.[lettre] then compte := !compte + 1
    done;
!compte;;

assert (nb_voy "scrabble" = 2) ;
assert (nb_voy "evaluation" = 6);;
````

````ocaml
nb_voy "pas un mot";;
````

**<p style = "color:purple">Exercice :</p>** 
1. Ecrire une fonction `max : 'a list -> 'a` qui renvoie l'élément maximum d'une liste.
1. Pourquoi obtient-on cette signature ? Quels sont les types possibles ?
<hr>

````ocaml
exception Liste_vide;;

let rec max = function
    | [] -> raise Liste_vide
    | [e] -> e
    | e::q -> let m = max q in
                if e > m then e
                else m
;;
````

````ocaml
max [7; -2; 12; -20; 0];;
````

````ocaml
max [];;
````

**<p style = "color:purple">Exercice :</p>** 
Ecrire une fonction `mem : 'a list -> 'a -> bool` telle que `mem liste e` renvoie `true` si l'élément `e` est présent dans la liste `mem` et `false` sinon.
<hr>

````ocaml
let rec mem liste elem = match liste with
    | [] -> false
    | e::q -> if e = elem then true else mem q elem
;;

assert (mem [3; 5; -1; 2] 2);
assert ( not (mem [3; 5; -1; 2] 8) );
assert (not (mem [] 4) );;
````

**<p style = "color:purple">Exercice :</p>** 
Ecrire une fonction `iter : int list -> int list` qui renvoie la liste passée en argument ou chaque élément a été incrémenté de 1.
<hr>

````ocaml
let rec iter = function
    | [] -> []
    | e::q -> (e + 1)::(iter q)
;;

assert ( (iter [6; 2; -5]) = [7; 3; -4]);;
````

**<span style = "color:red">/!\ Attention :</span>** La fonction `iter` définie ici n'a rien à voir avec la fonction `List.iter` du module `List`.

Deviner la signature puis ecrire une fonction `map` telle que `map f liste` renvoie la liste constituée des éléments de `liste` auxquels on a appliqué la fonction `f`.
<hr>

````ocaml
let rec map f = function
    | [] -> []
    | e::q -> (f e)::(map f q)
;;

assert ( (map (fun x -> x * x) [-3; 0; 4; -12]) = [9; 0; 16; 144]);;
````

**<p style = "color:purple">Exercice :</p>** 
1. Ecrire une fonction `split : 'a list -> 'a list * 'a list` qui partage une liste en deux listes de tailles égales (à 1 élément près).
1. Ecrire une fonction `split_p : int list -> int list * int list` qui partage une liste en deux listes : une contenant les éléments pairs, l'autre les éléments impairs.
<hr>

````ocaml
let rec split = function
    | [] -> ([], [])
    | [e] -> ([e], [])
    | e1::e2::q -> let q1, q2 = split q in (e1::q1, e2::q2)
;;

split [1; 2; 3; 4; 5; 6; 7];;
split [1; 2; 3; 4; 5; 6];;
````

````ocaml
let rec split_p = function
    | [] -> ([], [])
    | e::q -> let p, i = split_p q in
                if e mod 2 = 0 then (e::p, i)
                else (p, e::i)
;;

split_p [4; 2; 5; 8; 3; 9];;
````

# Accumulateur
**<p style = "color:purple">Exercice :</p>** 
Réécrire une nouvelle fois une fonction `fibonacci` mais cette fois avec une fonction auxiliaire récursive `fib` utilisant des accumulateurs : on appelera `fib n a b` lorsqu'il reste `n` termes à calculer et que `a` et `b` sont les deux derniers termes calculés.
<hr>

````ocaml
let fibonacci n =
    let rec fib n a b = match n with
        | 0 | 1 -> a
        | _ -> fib (n-1) (a + b) a
    in fib n 1 1
;;
fibonacci 3;;
fibonacci 5;;
````

````ocaml
fibonacci 45;;
````

**<p style = "color:purple">Exercice :</p>** 
Ecrire une fonction `retourne : 'a list -> 'a list` qui renvoie la liste passée en argument renversée. On pourra également utiliser une fonction auxiliaire récursive avec accumulateur.
<hr>

````ocaml
let retourne liste = 
    let rec passe l1 l2 = match l1 with
        | [] -> l2
        | e::q -> passe q (e::l2)
    in passe liste []
;;

retourne [1; 2; 3; 4; 9; 8; 7; 6];;
````

# Diviseurs
**<p style = "color:purple">Exercice :</p>** 
1. Ecrire une fonction `diviseurs : int -> int list` qui renvoie la liste des diviseurs d'un entier.
1. Ecrire une fonction `premier : int -> bool` qui teste si un entier naturel est premier ou non. On lèvera une exception si l'entier passé en argument est négatif.
1. Ecrire une seconde version de cette fonction qui n'utilise pas la fonction `diviseurs` mais plutôt une fonction auxiliaire qui lèvera une exception dès qu'on trouve un diviseur.
1. Ecrire une fonction `liste_premiers : int -> int list` telle que `liste_premiers n` renvoie la liste des `n` plus petits nombres premiers. 
<hr>

````ocaml
(* 1 *)
let diviseurs n =
    let rec div k i = match i with (* donne la liste de diviseurs de k inférieurs à i *)
        | 1 -> [1]
        | _ -> let liste = div k (i-1) in
                                if k mod i = 0 then i::liste
                                else liste
    in div n n
;;

diviseurs 45;;
````

````ocaml
(* 2 *)
exception Negatif
let premier n =
    if n < 0 then raise Negatif
        else List.length (diviseurs n) = 2
;;
````

````ocaml
assert (premier 13);
assert (premier 2);
assert (not (premier 1) );
assert (not (premier 16) );;
premier (-2);;
````

````ocaml
(* 3 *)
exception Diviseur;;
let premier_bis n =
    let rec cherche_div k i = match i with (* cherche s'il y a un diviseur de k inférieur à i *)
        | 1 -> true
        | _ -> if k mod i = 0 then raise Diviseur
                else cherche_div k (i-1)
    in 
    try cherche_div n (n-1)
        with Diviseur | Division_by_zero -> false
;;

assert (premier_bis 13);
assert (not (premier_bis 8));
assert (premier_bis 2);
assert (not (premier_bis 1));;
````

````ocaml
(* 4 *)
let liste_premiers n = 
    let rec next_prime k p liste = match k with
    | 0 -> liste
    | _ -> let e = p + 1 in
                if premier_bis e then next_prime (k-1) e (e::liste)
                else next_prime k e liste
    in next_prime n 1 []
;;

liste_premiers 10;;
````
