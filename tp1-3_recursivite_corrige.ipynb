# Premières fonctions récursives
**<p style = "color:purple">Exercice :</p>** 
1. Définir une fonction récursive `somme : int -> int` telle que `somme n` détermine la somme des entiers de `1` à `n`.
1. Vérifier que le résultat obtenu est bien le même que celui obtenu avec la formule explicite (vue en classe de première) pour les 100 premiers entiers.
<hr>

```ocaml
let rec somme n =
    if n = 1 then 1
    else n + somme (n-1)
;;
```

```ocaml
for n = 1 to 100 do
    assert (somme n = (n * (n + 1) / 2))
done;;
```

**<p style = "color:purple">Exercice :</p>** 
1. Définir une fonction récursive `pow : int -> int -> int` telle que `pow x n` élève `x` à la puissance `n`.
1. Combien de multiplications sont nécessaires pour pour calculer $a^n$ à l'aide de cette fonction ($(a, n)\in \mathbb{N}^\star \times \mathbb{N}$)?
<hr>

```ocaml
let rec pow x n =
    if n = 0 then 1
    else x * (pow x (n-1))
;;
```

```ocaml
assert (pow 2 3 = 8);;
```

**<p style = "color:purple">Exercice :</p>** 
Définir une fonction récursive `puissance_2 : int -> int` telle que `puissance_2 n` renvoie le plus grand entier $p$ tel que $n = a\times 2^p$ avec $a\in\mathbb{N}$ (on supposera $n\neq 0$).
<hr>

```ocaml
let rec puissance_2 n = 
    if n mod 2 = 0 then 1 + puissance_2 (n/2)
    else 0
;;
```

```ocaml
assert (puissance_2 13 = 0);
assert (puissance_2 48 = 4);;
```

**<p style = "color:purple">Exercice :</p>** 
1. Définir une fonction `premier : int -> bool` telle que `premier n` renvoie un booléen indiquant si `n` est un nombre premier ou non.
1. Définir une seconde fonction `premier_bis : int -> bool` qui utilise cette fois une fonction auxiliaire récursive. C'est-à-dire qui utilise une fonction récursive déclarée localement, à l'intérieur de la fonction.
<hr>

```ocaml
let premier n = let rep = ref true in
    let stop = int_of_float (sqrt (float_of_int n)) in
    for i = 2 to stop do
        if n mod i = 0 then rep := false
    done;
!rep;;
```

```ocaml
assert (premier 13);
assert (premier 2);
assert (not (premier 6));;
```

```ocaml
let premier_bis n =
    let rec diviseurs_avant i =
        if i = 1 then false
        else if n mod i = 0 then true
        else diviseurs_avant (i-1)
    in
not (diviseurs_avant (n-1));;
```

```ocaml
assert (premier_bis 13);
assert (premier_bis 2);
assert (not (premier_bis 6));;
```

# Temps de vol de la suite de Syracuse
**<p style = "color:purple">Exercice :</p>** 
Ecrire une fonction `tps_vol : int -> int` telle que `tps_vol nb` renvoie le temps de vol de la suite de Syracuse, c'est-à-dire le premier entier $n$ tel que $u_n = 1$ où la suite $(u_n)$ est définie par $u_0 = nb$ et $u_{n+1} = \begin{cases} \displaystyle\frac{u_n}2 & \text{si $n$ est pair} \\ 3u_n+1 & \text{sinon} \end{cases}$, pour tout $n\in\mathbb{N}$.
<hr>

```ocaml
let rec tps_vol n =
    if n = 1 then 0
    else if n mod 2 = 0 then 1 + tps_vol (n/2)
    else 1 + tps_vol (3 * n + 1)
;;
```

```ocaml
assert (tps_vol 8 = 3);
assert (tps_vol 5 = 5);;
```

# Pyramide
**<p style = "color:purple">Exercice :</p>** 
On souhaite construire un pyramide à l'aide de pierres cubiques. Un pyramide à $n$ étages possède une base carrée constituée de $n$ pierres par côté. On note $P_n$ le nombre de pierres dont est constituée une pyramide à $n$ étages.
1. Exprimer, en fonction de n, la relation de récurrence entre $P_n$ et $P_{n+1}$.
1. Définir une fonction `pyramide : int -> int` telle que `pyramide n`renvoie le nombre de pierres nécessaires à la construction d'une pyramide à `n`étages.
<hr>

Pour tout $n\in \mathbb{N}$, $P_{n+1} = (n+1)^2 + P_n$.

```ocaml
let rec pyramide n =
    if n = 0 then 0
    else n * n + pyramide (n-1)
;;
```

```ocaml
assert (pyramide 2 = 5);
assert (pyramide 3 = 14);;
```

# Fibonacci
**<p style = "color:purple">Exercice :</p>** 
La suite de Fibonacci est définie de la façon suivante :
$$\begin{cases} F_0 = 1\\
F_1 = 1\\
F_{n+2} = F_n + F_{n+1},\quad \forall n \in \mathbb{N}\end{cases}$$
1. Définir une fonction récursive `fibo : int -> int` telle que `fibo n` renvoie le terme de rang `n` de la suite de Fibonacci.
1. Tester cette fonction sur des petites valeurs.
1. Tester `fibo 42`. Que remarque-t-on ?
1. Tester `fibo 45`. Est-ce raisonnable d'essayer avec une plus grande valeur ?
1. Une explication ?
<hr>

```ocaml
let rec fibo n =
    if (n = 0) || (n = 1) then 1
    else (fibo (n-1)) + (fibo (n-2))
;;
```

```ocaml
assert (fibo 4 = 5)
```

```ocaml
fibo 42;;
```

Environ 3 secondes pour exécuter `fibo 42`.

```ocaml
fibo 45;;
```

Environ 10 secondes pour exécuter `fibo 45`.

La complexité en temps de ce programme est exponentielle !

# Division euclidienne
**<p style = "color:purple">Exercice :</p>** 
1. Définir une fonction récursive `pgcd : int -> int -> int` qui détermine le pgcd de deux entiers naturels (non tous nuls).
1. Vérifier qu'on obtient bien le bon résultat à l'appel de `pgcd 13 256`, `pgcd 4 0` et `pgcd 0 5`.
1. Copier cette première fonction puis la modifier en une fonction `euclide : int -> int -> int` qui affiche en plus les différentes divisions euclidiennes effectuées pour obtenir ce résultat. On pourra utiliser les fonction `print_int`, `print_string`et `print_newline`.
<hr>

```ocaml
let rec pgcd a b = if b = 0 then a
    else let r = a mod b in
        pgcd b r
;;
```

```ocaml
assert (pgcd 13 256 = 1);
assert (pgcd 4 0 = 4);
assert (pgcd 0 5 = 5)
```

```ocaml
let rec euclide a b = if b = 0 then a
    else let r = a mod b in
        begin
            print_int a;
            print_string "="; 
            print_int (a / b);
            print_string "x";
            print_int b;
            print_string "+";
            print_int r;
            print_newline ();
            euclide b r
        end
;;
```

```ocaml
euclide 256 13
```
