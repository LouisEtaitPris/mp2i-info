## Variables

La déclaration de variables en Ocaml s'effectue grâce au mot-clé `let` comme ceci :

```ocaml
let a = 8;;
```

La sortie indique qu'on vient de créer une **variable** dont :
* le nom est `a`
* le **type** est `int` : c'est un entier relatif
* la **valeur** vaut `8`

**<span style = "color:red">/!\ Important :</span>** En Ocaml, les expressions se terminent par `;;`

Le choix du nom de variable est important, on utilise en règle générale des mots **sans espace** et **non accentués**, utiliser le symbole `_` (underscore) si besoin. Enfin, plusieurs variables peuvent être déclarées simultanément.

Par exemple, pour calculer la distance parcourue en 3 h à une vitesse moyenne de 87 km/h on peut écrire :

```ocaml
let duree, km_par_h = 3, 87;;
duree * km_par_h;;
```

Nous venons de définir des variables **globales**. Lorsqu'on souhaite qu'une variable soit **locale** (i.e. définie uniquement dans le cadre d'une expression) on utilise la construction `let ... in`.
Dans l'exemple précédent il aurait par exemple été possible d'écrire plutôt comme ceci :

```ocaml
let duree_bis, km_par_h = 4, 87 in
        duree_bis * km_par_h;;
```

```ocaml
duree_bis;;
```

**<span style = "color:red">Commentaire :</span>** La variable `duree_bis` est une variable **locale** qui n'existe que pour l'expression qui suit le `in`.

On peut bien entendu effectuer des calculs utilisant des variables avec les opérations élémentaires `+` `-` `*` `/`.

Attention en revanche, le symbole `/` correspond à une division euclidienne. On obtient par exemple :

```ocaml
let b = a / 3;;    (*on rappelle que la variable a vaut 8*)
b;;             (* 8 / 3 = ... *)
```

**Remarque :** Pour insérer des commentaires comme dans la cellule précédente on inscrit ce commentaire entre `(*` et `*)`

```ocaml
13 / 5;;
13 mod 5;;
```

**<span style = "color:red">/!\ Attention :</span>** la valeur d'une variable définie comme nous l'avons fait jusqu'à présent n'est pas modifiable en place. Pour avoir des variables modifiables, on utilise en plus le mot-clé `ref`. 

On obtient alors une variable modifiable en place appelée **référence**.
* Le changement de la valeur d'une référence s'effectue avec `:=`.
* Pour accéder à la valeur d'une référence on ajoute un `!` au nom de la variable.

Par exemple :

```ocaml
let c = ref 24;;
```

```ocaml
c := 12;;
```

On remarquera le type `unit` de cette instruction : la valeur de `c` est modifiée, mais aucune valeur n'est renvoyée.

On dit qu'il s'agit d'un **effet de bord**.

```ocaml
c := !c / 3;;
!c;;
```

**<span style = "color:red">/!\ Attention :</span>** Ne pas confondre `:=` et `=` !!

`=` sert à **tester** une égalité (l'équivalent de `==` en Python).

```ocaml
!c = 64;;
```

```ocaml
!c = 4;;
```

Les valeurs renvoyées, `false` (faux) et `true` (vrai), sont de type `bool` (booléen).

## Fonctions

Une fonction Ocaml se définie comme ceci : `let ma_fonction mon_argument = ...`

Par exemple, la fonction ci-dessous met un entier au carré :

```ocaml
let carre x = x * x;;
```

Après exécution on voit apparaître la **signature** de la fonction : la fonction `carre` prend un argument de type `int` et renvoie un `int`.

Ce type est **inféré** par le compilateur grâce à l'opérateur `*`.

Pour une fonction à plusieurs arguments on écrit par exemple : `let ma_fonction arg1 arg2 arg3 = ...`.

**<span style = "color:purple">Exercice :</span>** 

1. Écrire une fonction `racine` qui détermine la racine carrée d'un nombre.
1. Écrire une fonction `norme` telle que `norme x y` calcule la norme du vecteur $\vec{u}\begin{pmatrix}x\\ y\end{pmatrix}$
1. Écrire une fonction `discrimant` telle que `discriminant a b c` calcule le discriminant du trinôme $ax^2 +bx + c$.

```ocaml
(* 1 *)
let racine x = x ** 0.5;;
racine 100. = 10.;;
racine 2.;;
```

```ocaml
(* 2 *)
let norme x y = racine ( (x *. x) +. (y *. y) );;
norme 1. 0. = 1.;;
norme 0. (-1.) = 1.;;
norme (-1.) 1. = racine 2.;;
```

**<span style = "color:red">Commentaire :</span>** Les parenthèses autour par exemple de `x *. x` ne sont pas nécessaires mais facilitent grandement la compréhension du code.<br>
Celles autour de `-1.` sont nécessaires, sans elles `0. -1.` serait interprété comme la soustraction entre 0 et 1.

```ocaml
(* 3 *)
let discriminant a b c = (b *. b) -. 4. *. a *. c;;
discriminant 1. 2. 1. = 0.;;
discriminant 1. 0. (-1.) = 4.;;
discriminant 1. 3. 1. = 5.;;
```
