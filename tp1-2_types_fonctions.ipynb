### Géométrie

Le type `float*float` désigne un couple de flottants, et peut représenter un point `p` dans $\mathbb{R}^2$.

On peut récupérer les coordonnées de `p` avec la déclaration `let x, y = p` . Par exemple :

````ocaml
let p = (2.1, 3.7) in (* exemple de point *)
let x, y = p in (* récupération des coordonnées *)
x;; (* affichage de x *)
````
