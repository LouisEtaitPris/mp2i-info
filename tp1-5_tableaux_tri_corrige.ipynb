**<p style = "color:purple">Exercice :</p>** 
Ecrire une fonction `analyse : string -> int array` telle que `analyse phrase` renvoie un tableau contenant :
* à la position 0 le nombre de caractères de `phrase` qui ne sont pas des lettres
* aux positions 1 à 26 le nombre de caractères dans `phrase` correspondant à chacune des lettres de l'alphabet (majuscule ou minuscule)

*On pourra s'aider de la fonction int_of_char s'appuyant sur la table de correspondance ASCII.*
<hr>

```ocaml
let analyse phrase =
    let stop = String.length phrase - 1 in (* calcul de la taille avant, pour éviter de recalculer à chaque passage en boucle *)
    let tab = Array.make 27 0 in    (* création d'un tableau initialisé à 0 *)
    for i = 0 to stop do
        let car = int_of_char (phrase.[i]) in    (* code ascii calculé une unique fois *)
        if car >= 65 && car <= 90 then 
            let pos = car - 64 in
            tab.(pos) <- tab.(pos) + 1
        else if car >= 97 && car <= 122 then
            let pos = car - 96 in
            tab.(pos) <- tab.(pos) + 1
        else tab.(0) <- tab.(0) + 1
    done;
tab;;
```

```ocaml
analyse "ceci est une phrase"
```

**<p style = "color:purple">Exercice :</p>** 
Ecrire une fonction `maxi : 'a array -> 'a` qui renvoie l'élément maximum d'un tableau.
<hr>

```ocaml
let maxi tab = let max = ref tab.(0) in
    let stop = Array.length tab - 1 in
    for i = 1 to stop do
        if tab.(i) > !max then max := tab.(i)
    done;
!max;;
```

```ocaml
maxi [| 3; -7; 12; -5; 8 |]
```

**<p style = "color:purple">Exercice :</p>** 
Ecrire une fonction `tri_insertion : 'a array -> unit` qui trie un tableau en place en procédent par insertion : les éléments sont triés un à un en les déplaçant dans une zone du tableau considérée comme triée.

Quelle est la complexité dans le pire des cas ?
<hr>

```ocaml
let tri_insertion tab =
    let swap i j = let tmp = tab.(i) in
        tab.(i) <- tab.(j);
        tab.(j) <- tmp
    in
    let stop = Array.length tab - 1 in  (* calcul de la taille du tableau une unique fois *)
    for i = 1 to stop do
        let j = ref i in
        while !j > 0 && tab.(!j) < tab.(!j - 1) do
            swap (!j) (!j - 1);
            j := !j - 1
        done
    done
;;
```

```ocaml
let tableau = [|2; -3; 5; 7; 2 |];;

tri_insertion tableau;;

tableau;;
```

Dans le pire cas, le tableau est trié dans l'ordre décroissant. Chaque élément est donc comparé à tous ceux qui le précèdent pour le positionner à la première position du tableau.

On effectue alors $\displaystyle\sum_{k=2}^n k - 1 = \sum_{m=1}^{n-1} m = \frac{(n-1)n}2$ comparaisons : la complexité, dans le pire cas, est quadratique (ainsi qu'en moyenne d'ailleurs).

Cet algorithme de tri reste néanmoins efficace sur un faible nombre de données ou sur des données presque triées (complexité linéaire dans le meilleur cas).

**<p style = "color:purple">Exercice :</p>** 
Ecrire une fonction `quick_sort : 'a array -> unit` qui trie un tableau en place selon la méthode suivante :
* on utilise un fonction auxiliaire qui trie entre deux indices i et j uniquement
* le tri qui s'effectue de manière récursive consiste à :
    * choisir un élément pivot
    * trier les éléments en mettant à gauche les éléments inférieurs au pivot, à droite ceux qui lui sont supérieurs
    * trier la partie gauche et la partie droite

Quelle est la complexité dans le pire des cas ?
<hr>

```ocaml
let quick_sort tab =
    let swap i j = let tmp = tab.(i) in    (* swap i j échange les valeurs aux indices i et j *)
        tab.(i) <- tab.(j);
        tab.(j) <- tmp
    in
    let rec tri i j = (* cette fonction trie tab entre les indices i et j *)
        if i < j then
            begin
                let p = tab.(i) in    (* on choisit un pivot *)
                let debut = ref (i + 1) and fin = ref j in  (* on initialise la zone des éléments non testés *)
                for k = i + 1 to j do
                    if tab.(!debut) > p then    (* on compare le premier élément de la zone avec le pivot *)
                        begin
                            swap !debut !fin;  (* chaque élément supérieur au pivot est placé à la fin de la zone *)
                            fin := !fin - 1
                        end        (* begin-end ou () obligatoires car if-then prioritaire par rapport au ; *)
                    else debut := !debut + 1
                done;
                swap i !fin;    (* le pivot est correctement positionné *)
                tri i (!fin - 1);
                tri (!fin + 1) j
            end
    in tri 0 (Array.length tab - 1)
```

```ocaml
let tableau = [|2; -3; 5; 7; 2 |];;

quick_sort tableau;;

tableau;;
```

* Le tri de k éléments nécessite de comparer k - 1 éléments au pivot.
* Dans le pire cas, le pivot sélectionné correspond toujours à l'élément le plus petit, il reste donc k - 1 éléments à trier.
* Pour trier n éléments on effectue donc dans le pire cas $\displaystyle\sum_{k = 2}^n k - 1 = \sum_{m = 1}^{n-1} m = \dfrac{(n-1)n}2$ comparaisons.

On obtient une complexité quadratique dans le pire cas.

Bonne nouvelle, en moyenne la complexité est de l'ordre de $n\log n$, qui est la complexité optimale d'un tri par comparaisons.

**<p style = "color:purple">Exercice :</p>** 
1. Récupérer dans le TP précédent la fonction `split : 'a list -> 'a list * 'a list` qui partage une liste en deux listes de tailles égales (à 1 élément près).
1. Ecrire une fonction `tri_fusion : 'a list -> 'a list` qui renvoie une liste triée en procédent récursivement comme ceci :
    * on partage la liste en deux listes
    * on trie chacune des deux listes
    * on fusionne les deux listes triées en une seule, toujours triée
<hr>

```ocaml
let rec tri_fusion liste =
    let rec split = function
        | [] -> ([], [])
        | [e] -> ([e], [])
        | e1::e2::q -> let q1, q2 = split q in (e1::q1, e2::q2)
    in
    let rec fusion_triee = function
        | [], [] -> []
        | l1, [] -> l1
        | [], l2 -> l2
        | e1::q1, e2::q2 -> if e1 < e2 then e1::(fusion_triee (q1, e2::q2) )
                            else e2::(fusion_triee (e1::q1, q2) )
    in
    match liste with
        | [] | [_] -> liste (* cas de base : les listes qu'on ne peut plus couper en deux sont celles de 0 ou 1 élément *)
        | _ -> let l1, l2 = split liste in
                    fusion_triee (tri_fusion l1, tri_fusion l2)
;;
    
tri_fusion [5; -6; 16; 20; 4; 1; 7];;
```

Le tri fusion, qui utilise une technique "diviser pour régner" a une complexité en $n\log n$, ce qui est optimal.
